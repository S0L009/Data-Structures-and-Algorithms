# Pseudo Code for Kruskal Algorithm



```python
def Helper(x,graph):
    if x[0] in graph.keys():graph[x[0]].append(x[1])
    else:
        graph.update({x[0]:[x[1]]})

    if x[1] in graph.keys():graph[x[1]].append(x[0])
    else:
        graph.update({x[1]:[x[0]]})
    return

def Cycle(st,graph,prev,prv):
    if prev.count(prev[0])>1:
        return True
    for i in graph[st]:
        if i!=prv:
            prev.append(i)
            a=Cycle(i,graph,prev[:],st)
            if a:
                return a
            prev.pop(-1)
    return False

def Krusk(ip,summ):
    edges=[]
    weights=[]
    for i in ip.keys():
        for j in ip[i]:
            lt=[i,j[0]]
            lt.sort()
            if lt not in edges:
                edges.append(lt)
                weights.append(j[1])

    for i in range(len(edges)):
        edges[i].append(weights[i])
    edges=sorted(edges,key=lambda i:i[2])
    print(edges)  #edges in index 0, 1 and weight of that edge in index 2
    graph={}
    for i in edges:
        Helper(i[:2],graph)
        if Cycle(i[0],graph,[i[0]],''):
            graph[i[0]].pop(-1)
            graph[i[1]].pop(-1)
        else:summ+=i[2]
    print(f'Total cost {summ}')
    return graph

inputt={
    1:[[6,10],[2,28]],
    2:[[1,28],[3,16],[7,14]],
    3:[[2,16],[4,12]],
    4:[[3,12],[7,18],[5,22]],
    5:[[4,22],[7,24],[6,25]],
    6:[[5,25],[1,10]],
    7:[[5,24],[4,18],[2,14]]
}

inputt=Krusk(inputt,0)
for i in inputt:
    print(i,'-',inputt[i])
```
