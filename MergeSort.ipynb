```python
class Merge_Sort:
    def initiate(self,Arr):
        return self.Merge(Arr,0,len(Arr)-1)
    
    def Combine(self,Left,Right):
        i=j=0
        lg=len(Left)
        while j<len(Right):
            if not i<lg:
                Left.append(Right[j])
                j+=1
                i=0
            elif Left[i]>Right[j]:
                Left.insert(i,Right[j])
                lg+=1
                j+=1
                i=-1
            i+=1
        return Left

    def Merge(self,Arr,l,r):
        if l==r:            #for one element
            return [Arr[l]]
        elif l+1==r:        #for 2 elements
            temp=Arr[l:r+1]
            return temp if Arr[l]<Arr[r] else temp[::-1]
        mid=(l+r)//2
        Left=self.Merge(Arr,l,mid)
        Right=self.Merge(Arr,mid+1,r)
        return self.Combine(Left,Right)
```

```python
import numpy as np
Arr=np.random.randint(-100,100,size=(1,10)).tolist()[0]
print(f"Input -> {Arr}")

Merge=Merge_Sort()
op_merge=Merge.initiate(Arr)
print(f"Merge Sort Output -> {op_merge}")

Arr.sort()
print(f"Inbuilt Sort Output -> {Arr}")

print(f"Are both of them same => {'üòÅ hehe...' if Arr==op_merge else 'üò≠ Noooooo'}")
```
