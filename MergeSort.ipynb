```python
class Merge_Sort:
    def initiate(self,Arr):
        return self.Merge(Arr,0,len(Arr)-1)
    
    def Combine(self,Left,Right):
        op = [0] * ( len(Left)+ len(Right) )
        l, r, k = 0, 0, 0
        while l < len(Left) and r < len(Right):
            if Left[l] < Right[r]:
                op[k] = Left[l]
                l += 1
            else:
                op[k] = Right[r]
                r += 1
            k += 1
        while l < len(Left):
            op[k] = Left[l]
            l += 1
            k += 1
        while r < len(Right):
            op[k] = Right[r]
            r += 1
            k += 1
        return op

    def Merge(self,Arr,l,r):
        if l==r:            #for one element
            return [Arr[l]]
        elif l+1==r:        #for 2 elements
            temp=Arr[l:r+1]
            return temp if Arr[l]<Arr[r] else temp[::-1]
        mid=(l+r)//2
        Left=self.Merge(Arr,l,mid)
        Right=self.Merge(Arr,mid+1,r)
        return self.Combine(Left,Right)
```

```python
merge=Merge_Sort()
merge.initiate([4,1,3,9,7])
```

```python
import numpy as np
Arr=np.random.randint(-10000,10000,size=(1,10000)).tolist()[0]
print(f"Input -> {Arr}\n")
import time
Merge=Merge_Sort()

st=time.time()
op_merge=Merge.initiate(Arr.copy())

print(f" time for My merge sort - {time.time()-st}\n")

print(f"Merge Sort Output -> {op_merge[:10]}  (first 10 elts)\n")
st=time.time()
Arr.sort()
print(f" time for inbuilt sort - {time.time()-st}\n")

print(f"Inbuilt Sort Output -> {Arr[:10]} (first 10 elts)")

print(f"Are both of them same => {'Yes  hehe...' if Arr[:]==op_merge[:] else ' Noooooo'}")
```
