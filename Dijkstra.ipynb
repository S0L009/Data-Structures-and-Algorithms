# Djikstra

```python
inputt={
'1':[['2',2],['3',4]],
'2':[['3',1],['4',7]],
'3':[['5',3]],
'4':[['6',1]],
'5':[['4',2],['6',5]],
'6':None
}

distance={'1':0,'2':99999,'3':99999,'4':99999,'5':99999,'6':99999}   #initial weights or distance
start='1'
```

```python
def Djnew(ip,start,distance,output_path):
    while ip[start]:
        mn=['',99999]
        for i in ip[start]:
            if distance[start]+i[1]<distance[i[0]]:
                distance[i[0]]=distance[start]+i[1]
                if mn[1]>distance[i[0]]:
                    mn[0]=i[0]
                    mn[1]=distance[i[0]]
        output_path.append(mn[0])
        start=output_path[-1]
    return output_path

print(Djnew(inputt,start,distance,[start]))
```
